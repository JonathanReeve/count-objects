```python
import json
import pandas as pd
from plotly import express as px
import statsmodels as sm
```

```python
with open('artifacts.jsonl') as f: 
    data = f.readlines()
```

```python
parsed = {}
for line in data:
    dataDict = json.loads(line)
    key = list(dataDict.keys())[0]
    parsed[key[8:40]] = dataDict[key]
```

```python
df = pd.DataFrame(parsed).T
px.line(df['structure.n.01'])
```

```python
df['year'] = df.index.str.slice(0,4).astype(int)
```

```python
subset = df[df.mean().sort_values(ascending=False).head(5).index]
```

```python
px.line(subset)
```
